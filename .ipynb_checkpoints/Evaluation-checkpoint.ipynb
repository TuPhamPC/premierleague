```python
import matplotlib.pyplot as plt
import numpy as np

# Chỉ số đánh giá
labels = ["Accuracy", "Precision", "Recall", "F1-score"]
knn_scores = [0.604, 0.512, 0.241, 0.328]
rf_scores = [0.647, 0.601, 0.355, 0.446]
svm_scores = [0.617, 0.605, 0.127, 0.211]
nn_scores = [0.624, 0.571, 0.247, 0.344]

x = np.arange(len(labels))
width = 0.15

# Vẽ biểu đồ cột
fig, ax = plt.subplots(figsize=(10, 6))
knn_bar = ax.bar(x - 1.5*width, knn_scores, width, label='kNN')
rf_bar = ax.bar(x - 0.5*width, rf_scores, width, label='Random Forest')
svm_bar = ax.bar(x + 0.5*width, svm_scores, width, label='SVM')
nn_bar = ax.bar(x + 1.5*width, nn_scores, width, label='Neural Network')

# Định dạng biểu đồ
ax.set_ylabel('Scores')
ax.set_title('Model Comparison')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Hiển thị giá trị trên các cột
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('%.2f' % height,
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(knn_bar)
autolabel(rf_bar)
autolabel(svm_bar)
autolabel(nn_bar)

plt.ylim(0, 1)  # Giới hạn trục y từ 0 đến 1
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

cv_accuracy = {
    'Neural Network': [62.3, 63, 61.9, 66.4, 63.7],  
    'Random Forest': [60.9, 64.4, 65.2, 59.9, 63.1],  
    'SVM': [61.5, 62.6, 61.7, 63.7, 61.4],  
    'kNN': [62.4, 62.4, 62.5, 61.1, 61.7]  
}

plt.figure(figsize=(8, 6))

for model, scores in cv_accuracy.items():
    plt.plot(range(1, 6), scores, marker='o', linestyle='-', label=model)  # Sử dụng range từ 1 đến 6

plt.xlabel('Fold')
plt.ylabel('Cross-Validation Accuracy Score')
plt.title('Cross-Validation Accuracy Score for Different Models')
plt.xticks(np.arange(1, 6))
plt.ylim(60, 70)
plt.legend()
plt.grid(True)
plt.show()

```
